```python
import pandas as pd
```

```python
def get_pop(df):
    df = pd.read_excel(r'https://www.abs.gov.au/statistics/people/population/national-state-and-territory-population/jun-2022/31010do002_202206.xlsx',
                     sheet_name='Table_7',
                   skiprows=list(range(0, 4)) + list(range(5, 222)) + list(range(329, 332)),
                  )
    
    df = df.rename(columns={'Age (years)':'age',
                                'New South Wales':'NSW',
                                'Victoria':'VIC',
                                'Queensland':'QLD',
                                'South Australia':'SA',
                                'Northern Territory':'NT',
                                'Australian Capital Territory':'ACT',
                                'Tasmania':'TAS',
                                'Western Australia':'WA',
                               'Australia':'AUS'})
    mask = df['age'].str.contains('[^\d]', regex=True, na=False)
    df = df[mask]
    
    df = pd.melt(df, id_vars=['age'], value_vars=['NSW','VIC','QLD','SA','WA','TAS','NT','ACT','AUS'])
    
    df.rename(columns={'age':'age_group',
                  'variable': 'state',
                  'value':'persons'},
         inplace=True)
    
    over_70 = df.loc[df['age_group'].isin(['70-74', '75-79', '80-84', '85-89', '90-94', '95-99','100 and over'])]
    over_70_row = pd.DataFrame(over_70.groupby(['state'])['persons'].sum())
    over_70_row['age_group'] = '70+'
    over_70_row.reset_index(inplace=True)
    
    df = pd.concat([df, over_70_row], axis=0).reset_index(drop=True)
    df = df[df['age_group'].isin(['70-74', '75-79', '80-84', '85-89', '90-94', '95-99','100 and over', 'All ages']) == False]
    df.sort_values(by=['state', 'age_group'])
    
    return df

    
```

```python
df = pd.DataFrame()
get_pop(df)
```
